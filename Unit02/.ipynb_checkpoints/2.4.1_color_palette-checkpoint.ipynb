```python
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')
                        
current_palette = sns.color_palette() 
sns.palplot(current_palette)
```

## seaborn 提供的基本 6 種調色板 

```python
theme_list=['deep', 'muted', 'pastel','bright', 'dark', 'colorblind'] 
for i in theme_list: 
    sns.palplot(sns.color_palette(i))
```

## HLS 多色調色板
* sns.color_palette(色彩空間名稱, 顏色數量)
* sns.hls_palette(顏色數量 n_colors, h, 亮度 l, 飽和度 s) # h,l,s 三個參數值皆介於 0 - 1 之間

```python
sns.color_palette("hls", 25)
```

```python
data = np.random.normal(size=(20, 12)) + np.arange(12) / 2  
'''
np.random.normal(size=(20, 12)): 產生一個 20 × 12 的矩陣，每個值皆來自 標準常態分布 N(0,1)
np.arange(12) → [0,1,2,...,11] 這是一個長度為 12 的向量
除以2 → [0,0.5,1,1.5,...,5.5]
** NumPy 會自動做 broadcasting：把這個長度 12 的向量「加到每一列」
'''
sns.boxplot(data=data,palette=sns.color_palette("hls", 12))
```

```python
#使用默認亮度和飽和度 
sns.palplot(sns.hls_palette(3))
```

```python
sns.palplot(sns.hls_palette(3, l=0.8, s=0.8))
```

## 連續性色板（漸層）

## 色調線性變換 sns.cubehelix_palette(n_colors, start, rot...)

```python
sns.palplot(sns.color_palette()) 
sns.palplot(sns.color_palette("cubehelix"))
```

```python
sns.palplot(sns.cubehelix_palette(8, start=0.8, rot=-0.5))
sns.palplot(sns.cubehelix_palette(8, start=0.1, rot=-0.8))
```

## 調用訂製連續調色板

```python
sns.palplot(sns.light_palette("green")) 
sns.palplot(sns.dark_palette("green"))
```

```python
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=600).T
'''
多變量常態分布（資料生成）
mean = [0, 0]: x 和 y 的平均數都是 0
covariance matrix  = [[1, -.5], [-.5, 1]]: Var(x) = 1, Var(y) = 1, Cov(x, y) = -0.5 --➜ x 和 y 有「負相關」
'''
pal = sns.dark_palette("green", as_cmap=True) # as_cmap=True → 給 KDE 用（連續色階）
sns.kdeplot(x = x, y = y, cmap=pal);
```

```python
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=600).T 
pal = sns.cubehelix_palette(light=1, as_cmap=True) 
sns.kdeplot(x = x, y = y, cmap=pal, shade=True)
```
