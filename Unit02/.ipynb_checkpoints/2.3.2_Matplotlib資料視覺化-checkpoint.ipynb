# `Matplotlib` 資料視覺化

`matplotlib` 幾乎是標準 Python 畫圖套件了! 在有 `matplotlib` 之前, Python 要畫圖不那麼方便, 和 Python 很多套件一樣, 有許多方案, 但各家有不同的優缺點, 也沒有一套是大家都在用的。

而 `matplotlib` 仿 Matlab 式的畫圖方式, 讓很多人很快入手、並且功能相當完整。原作者是 John D. Hunter, 和很多 Python 的套件作者一樣, 他有博士學位。非常令人遺憾的是他在 2012 年因大腸直腸癌治療併發症過逝, 過逝時才 44 歲!


```python
# 繪製的圖表（Figure）會直接顯示在程式碼儲存格（Cell）的下方，不需要每次都寫 plt.show()
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 1 畫個函數

前面說過 `plt.plot` 的基本用法是

    plt.plot(X,Y)

其中 X, Y 是分別是點 x 座標, y 座標形的的 list 或 array。

### 【重點】標準函數畫法

函數畫圖其實就是描點法, 而且再次用到陣列導向計算的概念。

我們先牛刀小試, 畫個 $\sin$ 的圖。

$$f(x) = \sin(x)$$

先來設個 x 的範圍, 就 -10 到 10, 給 100 個點。

```python
x = np.linspace(-10, 10, 100)
```

y 自然是 $\sin(x)$。

```python
y = np.sin(x)
```

畫圖!

```python
plt.plot(x,y)
```

### 【練習】函數畫圖

這次我們來畫:

$$f(x) = \frac{\sin(5x)}{1+x^2}$$

```python
x = np.linspace(0,10,200)
y = np.sin(5*x) / (1 + x**2)
plt.plot(x,y)
```

## 2 快速改變風格

雖然 `matplotlib` 在很辛苦的調校之後也是可以畫美美的圖, 不過它比較是「讓我們快速看到結果」的套件。所以它有些快速基本修飾功能。

### 【技巧】快速改變顏色

### 【技巧】快速改變線條風格

| 參數 | 說明 |

## 3 基本的修飾

| 參數 | 說明 |

## 4 如果畫個參數式圖形

其實參數式圖形基本上也是一堆 x, y 的座標, 只是 x(t), y(t) 都是 t 的函數, 我們可以想成 t 是時間。

假設我們的

$$ -2\pi \leq t \leq 2\pi$$

要畫個半徑是 r 的圓, x(t), y(t) 就是這樣:

* $x(t) = r\cos(t)$
* $y(t) = r\sin(t)$

我們來試試事情是不是如我們想的這樣...

```python
r = 3
```

```python
t = np.linspace(-2*np.pi, 2*np.pi, 200)
```

```python
x = r * np.cos(t)
y = r * np.sin(t)
plt.plot(x,y, lw = 3)
```

真的成功了... 等等, `matplotlib` 怎麼畫成橢圓呢? 我們來調整一下。

```python
ax = plt.gca() # gca = Get Current Axes（獲取當前坐標軸）
ax.set_aspect("equal") # Set Aspect Ratio to "Equal"（將縱橫比設置為「相等」）這會強制讓坐標軸的 X 軸單位長度與 Y 軸單位長度在物理顯示上相等。
plt.plot(x,y, lw = 3)
```

再來我們可以想一下, 如果 r 不是固定的呢? 也就是它會變長變短, 是不是會畫出很多特別的圖呢? 其實很多特別的曲線不過就是找個試合的 r(t), 我們來隨意試試看。

```python
r = np.sin(5*t)
x = r*np.cos(t)
y = r*np.sin(t)

ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x,y,lw=3)
```

是不是很有意思呢? 你也可以自己試試不同的變化。

## 5 `subplot` 畫多個圖

我們每次畫圖的時候, `matplotlib` 就弄 1 個 figure 畫圖區出來, 裡面可以有很多子圖, 在 figure 裡叫 axes。目前我們都只有 1 個 figure 內含 1 張圖, 所以都不用設, 現在我想畫 4 張圖時。我們就要先想好「陣式」。

比如說 2x2 這樣排列的 4 張圖。

```python
x = np.linspace(0, 10, 100)

plt.subplot(2, 2, 1) # 分成 2x2 個圖，現在在畫第一個圖
plt.plot(x, np.sin(x), c = "r", lw = 3)

plt.subplot(2, 2, 2)
plt.plot(x, np.sin(3*x), c = "g", lw = 3)

plt.subplot(2, 2, 3)
plt.scatter(x, np.random.randn(100), c = "y", s = 50, alpha=0.5)

plt.subplot(2, 2, 4)
plt.bar(range(10), np.random.randint(0,30,10), fc = "c")
```

## 6 進階色彩*

#### 顏色表示法 1

    c = 'r' 

可以用 blue (b), green (g), red (r), cyan (c), magenta (m), yellow (y), black (k), white (w)


#### 顏色表示法 2

用一個 0 到 1 的數字表灰階, 越大越白。

## 7 Marker*

所謂 'marker' 就是原來的點, 我們特別標示出來, 為了不要讓圖形太複雜, 我們新換個點比較少的例子。

```python
x = range(20)
y = np.random.randn(20) # randn: 產生標準常態分配的資料，第一個參數是一維陣列的尺寸
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, marker = "s")
```

#### marker 可以設的參數

| 參數 | 說明 |
|---|---|
| `marker` | marker 的風格 |
| `markeredgecolor` (`mec`) | 邊線顏色 |
| `markeredgewidth` (`mew`) | 邊線寬度 |
| `markerfacecolor` (`mfc`) | marker 的顏色 |
| `markerfacecoloralt` (`mfcalt`) | marker 替換色 |
| `markersize` (`ms`) | marker 大小 |
| `markevery` | 隔多少畫一個 marker |

```python
plt.plot(x, y, c = '#00a676', lw = 5, marker = "s", mfc = "y", mec = "#00b808", mew = 3, ms = 15)
```

比較奇特的參數是 `markevery`, 這是說我們每多少筆資料要畫一個 marker。為了示範, 我們拿之前的例子來看看。

```python
x = np.linspace(0,10,200)
y = np.sin(5*x) / (1 + x**2)
plt.plot(x,y, lw = 3, marker = "o", ms = 10)
```

後面的完完全全連在一起了, 我們現在隔 10 個畫一個 marker 試試。

```python
plt.plot(x,y, lw = 3, marker = "o", ms = 10, markevery =10)
```

## 9 畫圖區的設定*

#### 說明文字的標示

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y = np.sin(x)
```

```python
plt.title("My lovely sin function")
plt.xlabel('x-axes')
plt.ylabel('values')
plt.plot(x, y, lw=3)
```

#### 邊界的設定

```python
plt.xlim(-8,8) # xlim = x limit 
plt.ylim(-1.2,1.2)
plt.plot(x,y,lw=3)
```

#### xticks: 改變 x 軸標籤

```python
xv = np.linspace(0, 2*np.pi, 100)
yv = np.sin(xv)
plt.plot(xv,yv,lw=3)
```

```python
xv = np.linspace(0, 2*np.pi, 100)
yv = np.sin(xv)
plt.plot(xv,yv,lw=3)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']) # $ $: 是為了讓中間的字變成數學符號風格，如：𝝿, 𝜇, 𝛃
```

# [練習]自己生成測試數據，X軸xticks對應到y軸數值

例如: 1,2,3,4,5 對應X軸 '甲','乙'...

```python
x1 = np.linspace(0, 10, 50)
y1 = x1**2 +100
plt.plot(x1, y1, c='#7fb069')
plt.xticks([0,5,10],["low", "medium", "high"])
```

#### 圖例 legend: 要先在每個圖後面加 label, 最後記得畫一個 plt.legend()

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y = np.sin(x)
plt.xlim(-7,7)
plt.ylim(-1.2,1.2)
plt.plot(x, y, lw=3, label='$\sin$')
plt.plot(x, np.cos(x), lw=3, label='$\cos$')
plt.legend()
```

可用 `loc` 去設圖例的位置, 依 1, 2, 3, ... 表示。

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y = np.sin(x)
plt.xlim(-7,7)
plt.ylim(-1.2,1.2)
plt.plot(x, y, lw=3, label='$\sin$')
plt.plot(x, np.cos(x), lw=3, label='$\cos$')
plt.legend(loc = 1)
```

## 10 取得現在工作中 axes*

我們有時要設 axes 的背景啦等等的資訊。這時就要取得現在工作中的 axes。這一般有兩種方式, 第一種是設 `subplot` 時可以取得:

    fig, ax = plt.subplot()
    
另一種是用 `gca` 函數: *gca = Get Current Axes（獲取當前坐標軸）*

    ax = plt.gca() 

```python
ax = plt.gca() 
ax.set_facecolor('#69b8bb')
ax.set_xlim(-7,7)
ax.set_ylim(-1.2,1.2)
plt.plot(x,y,lw=5,c='white')
```

#### 移動 x, y 座標軸

```python
ax = plt.gca()
ax.set_facecolor('#69b8bb')
ax.set_xlim(-7,7)
ax.set_ylim(-1.2,1.2)

# 把上面和右邊的邊線拿掉(改成透明的)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 把下面和左邊的邊線改到中央的位置
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

plt.plot(x,y,lw=5,c='white')
```

## 12 `seaborn` 大救星

雖然 `matplotlib` 幾乎什麼都可以調, 但有些簡單的調整卻要花費很多時間。於是有了許多可以協助 `matplotlib` 變美、變方便的套件出現。最近有個很有名的叫 `seaborn`, 儼然有另一個標準套件的態式。

最重要的是 `seaborn` 和我們未來要介紹的 `pandas` 配合得非常好, 雖然熊貓和海生的聽來就沒什麼關係...

我們先來個標準 (又來標準縮寫) 讀入 `seaborn` 的方法。

```python
import seaborn as sns
```

然後像平常畫個圖。

```python
sns.set(color_codes = True)
```

```python
x = np.linspace(-10,10,200)
y = np.sinc(x)

plt.plot(x,y) # 因為前面曾經叫出可愛風 xkcd，所以之後畫的圖風格都會是這樣
```

```python
# 想要換回原本的風格時
plt.rcParams.update(save_state)

plt.plot(x, y)
```
