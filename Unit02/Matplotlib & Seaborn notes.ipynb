# **Matplotlib**
- 與 Numpy 結合去做視覺畫的效果

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 繪製基本圖表

### 折線圖

```python
data = [-1, -4.3, 15, 21, 31]
plt.plot(data)
plt.show()
```

```python
plt.plot(data, "o--b")
plt.show()
```

```python
# 繪製每日攝氏溫度的折線圖
days = range(1,8)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 31]
plt.plot(days, celsius)
plt.show()
```

```python
plt.plot(days, celsius, "--c^") # 更改外觀的字母順序不影響
plt.show()
```

```python
# 顯示圖表格線
plt.plot(days, celsius, "--c^") 
plt.grid(True)
plt.show()
```

```python
# 顯示 x y 軸標籤、標題
plt.plot(days, celsius, "--c^") 
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Temperatue in This Week")
plt.show()
```

```python
plt.plot(days, celsius, "--c^") 
print(plt.axis()) # 顯示軸的範圍
# 指定軸的自訂範圍
xmin, xmax, ymin, ymax = 0.5, 7.5, 15, 32.5
plt.axis([xmin, xmax, ymin, ymax])
plt.show()
```

#### 多條折線圖

```python
days = range(1,8)
celsius_min = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 31]
celsius_max = [27.6, 26.1, 22.5, 30.4, 29.5, 31.5, 32]

plt.plot(days, celsius_min, "--co",
         days, celsius_max, "r-o") 
plt.axis([0.5, 7.5, 15, 35])
plt.show()
```

```python
x = np.linspace(0,10,50) # numpy.linspace(start, stop, num=樣本數, endpoint=True「True 包含終始值，False 不包含」)等差數列
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, x, cosinus)
plt.show()
```

```python
plt.plot(x, sinus, "o-r", label = "sin(x)")
plt.plot(x, cosinus, "g--", label = "cos(x)")
'''
plt.xlabel("徑度")
plt.ylabel("振幅")
plt.title("sin 和 cos 波型")
'''
plt.legend(loc=1) # 圖例
plt.show()
```

### 散佈圖

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
plt.scatter(x,y)
plt.show()
```

### 長條圖 bar plots

```python
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
index = np.arange(len(labels))
ratings = [5.16, 5.73, 14.99, 3.17, 11.86, 4.45]
change = [1.12, 0.3, -1.69, 0.29, 3.41, -0.45]
```

```python
plt.bar(index, ratings)
plt.xticks(index, labels) # x軸每一欄的標籤
plt.ylabel("Usage") # 使用率
plt.title("Programming rating") # 程式設計使用率
plt.show()
```

```python
plt.bar(index, change)
plt.xticks(index, labels) # x軸每一欄的標籤
plt.ylabel("Change") 
plt.title("Programming rating") 
plt.show()
```

#### 水平長條圖

```python
plt.barh(index,change)
plt.xticks(index,labels)
plt.ylabel("Change")
plt.title("Programming rating")
plt.show()
```

### 直方圖 Histograms

```python
x = [21,42,23,4,5,26,77,88,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins = 5 # 分割次數
n, bins, patches = plt.hist(x,num_bins)
print(n)
print(bins)
print(patches)
plt.show()
```

### 圓餅圖 Pie Plots

```python
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
ratings = [5, 6, 15, 3, 12, 4]
```

```python
plt.pie(ratings, labels=labels)
plt.title("Programming ratings")
plt.axis("equal")
plt.show()
```

### 子圖表 Subplots
* plt.subplot(num_rows, num_cols, plot_num)

```python
x = np.linspace(0,10,50) 
sinus = np.sin(x)
cosinus = np.cos(x)
plt.subplot(2, 1, 1)
plt.plot(x, sinus, "r-o")
plt.subplot(2,1, 2)
plt.plot(x, cosinus, "g--")
plt.show()
```

### 多軸圖表

```python
x = np.linspace(0,10,50)
sinus = np.sin(x)
sinhs = np.sinh(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,sinus,'r-o')
ax2 = ax.twinx() # 共用 x 軸
ax2.plot(x,sinhs,'g--')
plt.show()
```

# **Seaborn**
- 與 Numpy & Pandas 結合去做視覺畫的效果

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(1,11) # x 軸的值
y = [3, 1, 8, 5, 4, 2, 4, 6, 3, 7] # y 軸的值

plt.title("change background") # 圖的標題
plt.xlabel("x axis") # x 軸的名稱
plt.ylabel("y axis") # y 軸的名稱
sns.set_style("whitegrid") # 白色網格背景
plt.bar(x, y) # 繪製長條圖
plt.show()
```

```python
x = np.arange(1,11) # x 軸的值
y = [3, 1, 8, 5, 4, 2, 4, 6, 3, 7] # y 軸的值


plt.title("change background") 
plt.xlabel("x axis") 
plt.ylabel("y axis") 
sns.despine() # 移除上放和右方的軸線
sns.set_palette("deep") # 設定調色盤
plt.bar(x, y) 
plt.show()
```
