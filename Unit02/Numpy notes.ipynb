## 建立 array

### 一維陣列
1. 先建立一個清單 list
2. 利用這個清單 List 與 元組 Tuple 去建立陣列
3. 之後可以運用索引的方式更改指定陣列的值

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)

# 要記得「先看 row, 再看 column」
arr2[0,0] = -10
print(arr2)
print(arr2[0,0], arr2[0,1], arr2[0,2])
print(arr2[1,0], arr2[1,1], arr2[1,2])
```

## 向量運算

```python
h = np.array([1,2,3])
print(h)

i = 5  # i 是個純量
print(i)
```

```python
# h1 = h + i
print(h)
h1 = h + i
print(h1)
h2 = h - i
print(h2)
h3 = h * i
print(h3)
h4 = h / i
print(h4)
```

```python
# 相同維度的向量才可以相加
j = np.arange(2, 5)
print("h =",h)
print("j =",j)
h5 = h + j
print(h5)
h6 = h - j
print(h6)
h7 = h * j
print(h7)
h8 = h / j
print(h8)
```

#### 點積運算 dot product: 把相對應的向量相乘後的結果全部加總起來

```python
print("h =",h)
print("j =",j)
print("h * j =", h7)

h9 = h.dot(j)
print(h9)
```

## 向量的長度: 使用 numpy 的 linalg module 的 norm method 處理

範例：假設今天住家位於座標（0,0），公園座標（1,3），商店（2,1），公司（4,4）

```python
# 住家到公園的距離
park = np.array([1,3])
distance_park = np.linalg.norm(park)
print(distance_park)
```

```python
# 從商店到公司的距離
store = np.array([2,1])
office = np.array([4,4])
store_office = office - store
print(store_office)
distance_store_office = np.linalg.norm(store_office)
print(distance_store_office)
```

## 陣列複製: copy()
* copy 在陣列中很重要！他可以保留原始值，不管我們後續對複製的陣列做什麼更動，都不影響原始陣列！！

```python
q = p.copy()
print(q)
print(p)
q[1] = -2
print(q)
print(p)
```

```python
id(p)
```

```python
id(q)
```

```python
# 直接用等號指派變數的話，我們後續只要對其中一個變數做更動，另一個也會一起變更，因為他們在程式中其實是被定義為同個變數
r = p
print(r)
print(p)
r[1] = 100
print(r)
print(p)
```

### 指定不同維度的計算

```python
y = np.arange(0, 10).reshape(2,5)
print(y)
```

```python
print(np.sum(y))
```

```python
print(np.sum(y, axis = 0)) # axis = 0 縱向求合
print(np.sum(y, axis = 1)) # axis = 1 橫向求合
```

```python
# mean 算術平均
print(np.mean(y, axis = 0)) 

# average 算加權平均
weight = [0.3, 0.7]
print(np.average(y, axis =0, weights=weight))
```
